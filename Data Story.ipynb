### What is distribution of Prize?

We see from the below scatter plot that there are some listings whos price can be considered as outliers and thus we remove these outliers. After removing the outliers we see the actual distributions of price. We see that the distribution of price is between 2000 dollars to around 13000 dollars.

```python
count=nd['price']
count
```

```python
plt.figure(figsize=(15,8))
plt.scatter(range(nd.shape[0]),nd['price'])
```

```python
ulimit = np.percentile(nd.price.values, 99)
nd['price'].ix[nd['price']>ulimit] = ulimit
```

```python
plt.figure(figsize=(15,8))
plt.scatter(range(nd.shape[0]),nd['price'])
```

### How is Price distributed with respect to the 3 categories?

```python
plt.figure(figsize=(15,8))
order = ['low', 'medium', 'high']
sns.stripplot(data['interest_level'],nd['price'],jitter=True,order=order)
plt.title("Price Vs Interest_level");
```

We see that the listings with high interest level have their prices between 1000k to at around 8000k. The lsitings with low interst levels have their prices between 1000k to more than 13k. It is strange that both high and low listing have common prices: ie between 1000k to around 8000k. Maybe looking at other variables like bedrooms and bathrooms will give us better answers.
